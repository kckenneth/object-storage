```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Feb 14 15:20:37 2018

@author: Kenneth Chen
"""

# ---------- Importing modules -------------------------

import sys
import time
from score_word import score
from itertools import permutations

start_time = time.time()

# ---------- Loading the dictionary corpus -------------
# dictionary with words whose characters up to 7 letters

with open("ppd.txt","r") as infile:
    raw_input = infile.readlines()
    data = [datum.strip('\n') for datum in raw_input]


# ---------- Main Script--------------------------------
    
def scrabble(rack_ori, letter=None, pos=None):
    """Cheating the scrabble game by available tiles with maximum score. \n
    Wild cards allowed are "*" and "?" for any number of wild cards."""

    rack = rack_ori.lower()

    # wild cards replacement with * for while loop, would do for any numbers of wild cards
    if "?" in rack:
        rack = rack.replace("?", "*")
        
    star_count = rack.count("*")
    
    # holder for all letters combination, eg 'boot', 'too', bot'
    scramble_list = []           
    for i in range(2, len(rack)+1):                          
        scramble_list.extend([''.join(p) for p in permutations(rack, i)])
      
    # wild card replacement function    
    def wild_card(scramble_list):
        """Replacing * with 26 alphabets, one at a time"""
        
        star_card_list = []
        for i in range(len(scramble_list)):
            for char in 'abcdefghijklmnopqrstuvwxyz':
                star_card_list.append(scramble_list[i].replace("*", char, 1))
        return list(set(scramble_list + star_card_list))
    
    while star_count >= 1:
        scramble_list = wild_card(scramble_list)
        star_count -= 1
        
        
    # ------- Scoring words if they're in the dictionary ----------
    
    # words_dict has scrabbled-words, eg 'cab', but not 'bac', to calculate scores in next step
    words_dict = list(set(scramble_list) & set(data))

    # words_list will have a tuple (score, word)
    words_list = []
    for item in words_dict:
        words_list.append(score(item, rack))
    
    words_list.sort(reverse= True)

    
    # ------- This is for extra credits + 10 ----------------------

    if letter != None:
        custom_word = []
        for i in range(len(words_list)):
            if letter in words_list[i][1]:
                custom_word.append(words_list[i])
        words_list = custom_word
    
    if pos != None:       
        custom_pos = []
        for i in range(len(words_list)):
            if len(words_list[i][1]) >= pos:
                custom_pos.append(words_list[i])
          
        custom = []
        for i in range(len(custom_pos)):
            for index, char in enumerate(custom_pos[i][1]):
                if index == pos-1 and char == letter:
                    custom.append(custom_pos[i])
        words_list = custom
    
    return words_list

# ---------- Reading argument (input) from the command line
    
if __name__ == '__main__': 
    if len(sys.argv) < 2 or sys.argv[1].isdigit():
        rack_ori = input("Please provide your scrabble rack: ")     
        result = scrabble(rack_ori)
    elif len(sys.argv) == 2:
        rack_ori = sys.argv[1]
        result = scrabble(rack_ori)
    elif len(sys.argv) == 3:
        rack_ori = sys.argv[1]
        letter = sys.argv[2].lower()
        result = scrabble(rack_ori, letter)
    elif len(sys.argv) == 4:
        rack_ori = sys.argv[1]
        letter = sys.argv[2].lower()
        pos = int(sys.argv[3])
        result = scrabble(rack_ori, letter, pos)
    else:
        print("Wrong input has been entered.")
       
    # -------- print out the final scrabble words ----------
    
    for item in result:
        print(*item, sep=' ')
    
    print("The total number of scrabble words for '{}' is {}".format(rack_ori, len(result)))
    print("The time taken to run the entire program is {}".format(time.time() - start_time))
    print("The highest scrabble words is: ", result[0])
```
